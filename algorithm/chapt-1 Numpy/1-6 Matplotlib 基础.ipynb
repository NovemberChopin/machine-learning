## matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlim(-5, 25) # 调节 x 轴的范围
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, label="cos(x)", color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.title("Welcome to the mechine learning")
```
